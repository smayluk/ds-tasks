# Поиск наиболее близкого имени участника

```python
# conda install -c conda-forge python-levenshtein
# или pip install python-c

import Levenshtein
```

```python
# что было написано в заявке

claim_list = [
    'Шехавцова Анна',
    'Гречихина Наталья',
    'Козлова Алена',
    'Груздева Алина',
    'Кущенко Анна',
    'Чистякова Анастасия'
]
```

```python
# результат расшифровки речи диктора

speech_recognition = [
    'кучменко она',
    'кущенко оксана',
    'груздь алина',
    'рычихина наталья',
    'шиховцева на',
    'чистова анастасия'
]
```

### С библиотекой Levenshtein

```python
Levenshtein.distance('Оксана', 'Анна')
```

```python
for recognized_name in speech_recognition:
    print(recognized_name)
```

```python
recognition_claim_relations = {}
for recognized_name in speech_recognition:
    recognition_claim_relations[recognized_name] = {}
    for real_name in claim_list:
        recognition_claim_relations[recognized_name][real_name] = Levenshtein.distance(recognized_name, real_name)
        #print('{},{},{}'.format(recognized_name, real_name, Levenshtein.distance(recognized_name, real_name)))
```

```python
for (i,recognized_name) in enumerate(recognition_claim_relations):
    min_levenshtein = sorted(recognition_claim_relations[recognized_name].items(), key = lambda x: x[1])[0]
    print('{} => {}'.format(recognized_name, min_levenshtein[0]))
```
